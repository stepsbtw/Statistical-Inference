# Estimadores
É o nome dado para **estatísticas** que usamos para estimar (aproximar) parâmetros.

Lembrando que:
- estatística -> função calculada em uma amostra.
- parâmetro -> número referente a população.

São denotados com o hat: image.png

Um estimador é classificado em relação a sua **Precisão** e **Viés**.

### Precisão
O quão próximo, em torno de θ, os valores do estimador irão se distribuir?

Calculado a partir da **Variância**.

image.png

- Nesse caso a precisão seria : 1 / Variância.

###Viés
Os valores do estimador estão concentrados em torno da média? Ou estão concentrados em outro ponto?

Calculado pela **diferença** entre o **valor esperado do estimador** e o valor do parâmetro real.

image.png
