# Teste ANOVA (ANalysis Of VAriance)
Usado para comparar **múltiplas** médias (três ou mais) e verificar se a diferença entre as médias amostrais é significante.

É um teste estatístico para determinar se as médias de vários grupos são iguais. Uma **generalização do t-teste** para os casos com mais de dois grupos.

Para cada elemento nas amostras:
- A variável independente é discreta
- A variável dependente é contínua

## Hipóteses
($H_0$) declara que não há diferença estatisticamente significativa entre as médias das populações. ($H_a$) declara que há diferença de ao menos um par de populações.

Com $k$ amostras:

- $H_0$: $\mu_1 = \mu_2 = \ldots = \mu_k$
- $H_a$: existem índices $i$ e $j$ tais que $\mu_i \neq \mu_j$ ($1 \leq i, j \leq k$)

## Distribuição F
A estatística de teste da ANOVA segue a **distribuição $F$** ($F$-distribution).

$X \sim F(d_1, d_2)$.

Valores sempre positivos.

## Ideia

A ideia central da ANOVA é "quebrar" a variância total dos dados em três componentes:
- Variação entre os grupos (devido ao fator de interesse)
- Variação dentro dos grupos (devido ao erro ou variabilidade aleatória)
- Graus de liberdade

Sendo:
- $k$, a quantidade de **grupos**
- $\overline{x}$: a média global (todas $k$ amostras)
- $n_1, n_2\ldots, n_k$: tamanho de cada grupo (amostra)
- $\overline{x}_1, \overline{x}_2, \ldots, \overline{x}_k$: média de cada grupo

$$
        \overline{x}_i = \frac{1}{n_i} \sum_{j=1}^{n_{i}} x_{ij}
$$

- $s_1^2, s_2^2, \ldots, s_k^2$: a variância de cada grupo
$$
s_i^2 = \frac{1}{n_i-1} \sum_{j=1}^{n_{i}}\left(x_{ij}-{\bar{x}}_{i}\right)^{2}
$$

- $N$ = tamanho total juntando as amostras

### Variância entre grupos (*between-group variability*)

$$
\frac{1}{k-1}\sum_{i=1}^{k}n_{i}({\bar{x}}_{i} - {\bar {x}})^{2} = \frac{n_1(\bar{x}_1 - \bar{x})^2 + \cdots + n_k(\bar{x}_k-\bar{x})^2}{k-1},
$$

### Variância dentro dos grupos (*within-group variability*)

$$
\frac{1}{N-k} \sum_{i=1}^{k}\sum_{j=1}^{n_{i}}\left(x_{ij}-{\bar{x}}_{i}\right)^{2} = \frac{(n_1-1)s_1^2 + \cdots + (n_k-1)s_k^2 }{N-k},
$$

## Estatística $F$
A razão entre as variâncias.
$$
F = \frac{\text{variabilidade inter-grupos}}{\text{variabilidade intra-grupos}}
$$

Os graus de liberdade da distribuição F são:
- $d_1 = k - 1$ (k grupos – 1)
- $d_2 = N - k$ (N total – k grupos)
