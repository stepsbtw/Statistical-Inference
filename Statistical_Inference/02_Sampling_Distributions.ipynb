# Sampling Distributions
- A Sampling Distribution models a statistic. It approximates a Parameter from the Population.

- In theory, a Sampling Distribution should represent all possible Simple Random Samples of that statistic.

( A Distribution models a function of probability f(x) - The probability of being value x )

# Empiric Sampling Distribution
- In practice, we choose a fixed number N of Samples to approximate that theorical Sampling Distribution. ( It should be Large enough to represent well )

(Sampling Distribution != Sample Distribution)

- The distribution of a Sample generally represent the proportions of the original Population Sample ( if large enough )

With that, the size of the Samples are important too. The larger they are, more it resembles the Population Distribution.

# Sampling Distribution of the Mean
- The statistic modeled is the Sample Mean x̅n, calculed with samples of size n.



```python
import numpy as np
import matplotlib.pyplot as plt

# parameters of the normal dist
mu = 0
sigma = 1
pop_size = 10**4
sample_size = 30
qt_samples = 1000

# generating the population
population = np.random.normal(mu, sigma, pop_size) # std normal dist

# creating the Sampling Distribution
sample_means = []

for _ in range(qt_samples):
  # generates 1 sample
  sample = np.random.choice(population, size = sample_size)
  sample_means.append(np.mean(sample))

# THE POPULATION
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.hist(population, bins = 30)
plt.title('Population')
plt.xlabel('Value')
plt.ylabel('Freq')

# THE SAMPLE DISTRIBUTION
plt.subplot(1,2,2)
plt.hist(sample_means, bins = 30)
plt.title("Sampling Distribution of the Means")
plt.xlabel("Mean of the Sample")
plt.ylabel("Freq")

plt.tight_layout()
plt.show()
```

It is aproximating really well the mean. But,
### Sampling Distribution of the Sample Mean (STD DEVIATION)
- It's clear that there is something wrong.

The CLT affirms that the std dev of the Sampling Dist of the Sample Mean
has μx => μ AND **σ²x = σ²/n**

- Which means that **σx = σ/sqrt(n)**, so we need to normalize it! The values became under-inflated.
