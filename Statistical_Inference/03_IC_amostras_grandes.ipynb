# Intervalo de Confiança para a Média (n>30)
Para amostras grandes, o TLC garante para a Distribuição Amostral da Média:
- mu_x = mu
- var_x = var/n

Como a proporção é uma média com apenas valores binários (Bernoulli), o TLC garante para a Distribuição Amostral da Proporção:
- mu_p_hat = p
- var_p_hat = p(1-p)/n

# Determinar o tamanho da amostra (n)

**Erro de estimação** ou **Margem de erro** ou **Erro Amostral** corresponde a metade do comprimento do intervalo de confiança.

Esse erro é inversamente proporcional ao tamanho da amostra, uma amostra maior, o erro amostral diminui.

Então dado o Erro Amostral, e um nível de confiança, posso determinar o n mínimo para cumprir tais "condicoes".
